# Project 2: Data Analysis and Machine Learning

# Import the Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (16,9)
```

# Analyzing the Results

```python
df_sample = pd.read_csv("sample_results.csv", index_col=0)
df_large = pd.read_csv("large_sample_results.csv", index_col=0)
df_top20 = pd.read_csv("Top20_large_sample_results.csv", index_col=0)
```

```python
plt.title("Small Sample Stock Universe Results")
for col in df_sample.columns[:-1]:
    plt.scatter(df_sample.index, df_sample[col], label=col)
plt.hlines(0.5014, df_sample.index[0], df_sample.index[-1], label="Accuracy Threshold", linestyles='dashed', colors='gold')
plt.hlines(0.5141, df_sample.index[0], df_sample.index[-1], label="Precision Threshold", linestyles='dashed', colors='silver')

plt.legend()
plt.show()
```

```python
plt.title("Large Sample Stock Universe Results")
for col in df_large.columns[:-1]:
    plt.scatter(df_large.index, df_large[col], label=col)
plt.hlines(0.5014, df_large.index[0], df_large.index[-1], label="Accuracy Threshold", linestyles='dashed', colors='gold')
plt.hlines(0.5141, df_large.index[0], df_large.index[-1], label="Precision Threshold", linestyles='dashed', colors='silver')

plt.legend()
plt.show()
```

```python
plt.title("Top 20 Large Sample Stock Universe Results")
for col in df_top20.columns[:-1]:
    plt.scatter(df_top20.index, df_top20[col], label=col)
plt.hlines(0.5014, df_top20.index[0], df_top20.index[-1], label="Accuracy Threshold", linestyles='dashed', colors='gold')
plt.hlines(0.5141, df_top20.index[0], df_top20.index[-1], label="Precision Threshold", linestyles='dashed', colors='silver')

plt.legend()
plt.show()
```
